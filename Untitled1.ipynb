```python
from PIL import Image, ImageOps, ImageFilter, ImageFont, ImageDraw
flowersImage = Image.open('./img/12.jpg')
flowersImage.thumbnail((800, 700))
flowersImage = flowersImage.convert('L')
flowersImage = flowersImage.filter(ImageFilter.FIND_EDGES)
flowersImage = flowersImage.point(lambda p: 255 - p)
flowersImage
display(flowersImage)

PotterImage = Image.open('./img/Garry.png')
PotterImage.thumbnail((500, 500))
PotterImage = PotterImage.convert("RGB")
r, g, b = PotterImage.split()
zeroed_band = r.point(lambda p: p * 0)
clist = ["r", "g", "b", "zeroed_band"]

import random

def rndm_text_color():
    return tuple(random.sample(range(1, 255), 3))

def createBandName(newBand):
    result_band_list = []
    for i in range(len(newBand)):
        if newBand[i] == "r":
            result_band_list.append(r)
        elif newBand[i] == "b":
            result_band_list.append(b)
        elif newBand[i] == "g":
            result_band_list.append(g)
        else:
            result_band_list.append(zeroed_band)
    return result_band_list

def createResImage(img, newBand, result_band_list):
    img = Image.merge("RGB", result_band_list)
    font_img = ImageFont.truetype("arial.ttf", size = 20)
    text_img = ""
    for k in range(len(newBand)):
        text_img += str(newBand[k])
        if k + 1 != len(newBand):
            text_img += " + "
    draw_image = ImageDraw.Draw(img)
    draw_image.rectangle((0,465,500,500), fill='black')
    draw_image = draw_image.text((10, 470), text_img, fill = rndm_text_color(), font = font_img)
    return img

def createCollage(img, col, row):
    newImg = Image.new('RGB', (img.width * col, img.height * row))
    for i in range(col):
        for j in range(row):
            newBand = random.choices(clist, k = 3)
            result_band_list = createBandName(newBand)
            img = createResImage(img, newBand, result_band_list)
            newImg.paste(img, (i * img.width, j * img.height))
    return newImg

createCollage(PotterImage, 3, 3).save('./img/HarryPotterCollage1.png')
createCollage(PotterImage, 3, 3).save('./img/HarryPotterCollage2.png')
createCollage(PotterImage, 3, 3).save('./img/HarryPotterCollage3.png')
imgs = Image.open('./img/HarryPotterCollage1.png')
imgs = Image.open('./img/HarryPotterCollage2.png')
imgs = Image.open('./img/HarryPotterCollage3.png')
display(imgs)
```
